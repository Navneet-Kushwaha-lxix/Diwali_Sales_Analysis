```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
# use encoding='unicode_escape' for unicode error
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax=sns.countplot(x='Gender',data=df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_by_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

bar_plot = sns.barplot(x='Gender', y='Amount', data=sales_by_gender)

# Add labels to the bars
#for bar in bar_plot.patches:
    #bar_plot.annotate(format(bar.get_height(), '.2f'),
                      #(bar.get_x() + bar.get_width() / 2,
                       #bar.get_height()), ha='center', va='center',
                      #size=12, xytext=(0, 8),
                      #textcoords='offset points')
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount', data=sales_age)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(7,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

## Conclusion:

- Count of female customer and their purchasing power is > male customer.
- Female customers between age group 26-35 have the maximum count and purchasing power followed by age group between 36-45.
- Unmarried female spends more than married female (especially between age of 26-35)
- Top 3 states by orders and purchasing power are U.P., Maharashtra and Karnataka.
- I.T. > Heathcare > Aviation } --> In order of, both by count and amount spent.
- Top 3 Category of Product by revenue are food, clothing and electronics.
